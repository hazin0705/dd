```python
import matplotlib.pyplot as plt

cat = plt.imread('/content/sample_data/cat.1.jpg')
dog = plt.imread('/content/sample_data/dog.1.jpg')
img = [cat,dog]
labels = ['cat','dog']

plt.figure(figsize =(5,3))
for i in range(len(img)):
  plt.subplot(1,2,i+1)
  plt.imshow(img[i])
  plt.xticks([])
  plt.yticks([])
  plt.xlabel(labels[i])
plt.show()
```

```python
import matplotlib.pyplot as plt
import os

imgs = os.listdir('dog')

plt.figure(figsize =(5,5))
for i in range(len(imgs)):
  plt.subplot(2,2,i+1)
  img = plt.imread(os.path.join('dog',imgs[i]))
  plt.imshow(img)
  plt.axis('off')
plt.show()
```

```python
import matplotlib.pyplot as plt
import random

dice1 = plt.imread("/content/dice/dice1.png")
dice2 = plt.imread("/content/dice/dice2.png")
dice3 = plt.imread("/content/dice/dice3.png")
dice4 = plt.imread("/content/dice/dice4.png")
dice5 = plt.imread("/content/dice/dice5.png")
dice6 = plt.imread("/content/dice/dice6.png")

dices = [dice1,dice2,dice3,dice4,dice5,dice6]

plt.figure(figsize =(5,5))
for i in range(2):
  plt.subplot(1,2,i+1)
  select = random.randint(0,len(dices)-1)
  plt.imshow(dices[select])
  plt.axis('off')
plt.show()
```
